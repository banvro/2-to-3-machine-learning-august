```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("data.csv")
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
```

```python
# a=data.loc("duration",axis=0)

x = data.iloc[:,0:1]
y = data.iloc[:,2:3]
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
obj = LinearRegression()
```

```python
obj
```

```python
obj.fit(x_train, y_train)
```

```python
y_test
```

```python

obj.predict(x_test.iloc[0].values.reshape(1, 1))
```

```python
m = obj.coef_
b = obj.intercept_

print(m , b)
# y = mx + b
```

```python
m * 200 + b
```
